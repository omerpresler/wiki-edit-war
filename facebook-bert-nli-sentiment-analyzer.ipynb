```python
from transformers import pipeline

# Load zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Candidate stance labels
labels = ["pro-israel", "neutral", "pro-palestine"]

# Wikipedia revision diffs (examples)
revisions = [
    "Israel has the right to defend itself against terrorism.",
    "The occupation of Palestinian land is widely condemned by the international community.",
    "A ceasefire was declared by both sides.",
    "The IDF launched an operation in Gaza following rocket fire.",
    "The settlers were evacuated under international pressure."
]

# Classify each revision
for rev in revisions:
    result = classifier(rev, candidate_labels=labels)
    top_label = result['labels'][0]
    print(f"Revision: {rev}")
    print(f"Top Label: {top_label} | Scores: {dict(zip(result['labels'], result['scores']))}\n")
```
